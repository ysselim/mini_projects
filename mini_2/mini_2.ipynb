# YELP SET-UP

```python
url = "https://api.yelp.com/v3/businesses/search?"

headers = {
    "Accept": "application/json",
    "Authorization": "Bearer " + yelp_key
}
parameters = {
    "location": "NYC",
    "term": "pizza",
    "sort_by": "rating",
    "limit": 50
    
    
}
response = requests.get(url, headers = headers, params = parameters)

data = response.json()
data
```

```python
all_pizzerias = []
for pizza in data['businesses']:
    id_pizzeria = pizza['id']
    
    url = f'https://api.yelp.com/v3/businesses/{id_pizzeria}'
    resp = requests.get(url=url, headers=headers)
    
    if resp.status_code == 200:
        pizzeria_data = resp.json()
        
        all_pizzerias.append(
            {
            'name': pizza['name'],
            'price_category': pizzeria_data['price'].strip() if pizzeria_data.get('price') else None,
            'rating': pizzeria_data['rating']*2,
            'api': 'yelp',
        })
```

```python
df_yelp = pd.DataFrame(all_pizzerias)
df_yelp['price_category'] = df_yelp['price_category'].apply(lambda x: len(x.strip()) if isinstance(x,str) else None)
```

```python
df_yelp
```

```python
df_yelp2 = df_yelp.copy()
df_yelp2 = df_yelp.rename(columns={"price_category": "price"})
df_yelp2.sort_values('rating', ascending = False)
```
